```python
import os
import re
import pandas as pd
from collections import defaultdict
import matplotlib.pyplot as plt

# File paths
file_paths = [
    'C:/Users/Dev/Documents/GitHub/Dona_project/Example_chat_data/WhatsApp_Chat_with_Donna_Patterson.txt',
    'C:/Users/Dev/Documents/GitHub/Dona_project/Example_chat_data/WhatsApp_Chat_with_Dr._Heather_Hanson.txt',
    'C:/Users/Dev/Documents/GitHub/Dona_project/Example_chat_data/WhatsApp_Chat_with_Jeffery_Hill_.txt',
    'C:/Users/Dev/Documents/GitHub/Dona_project/Example_chat_data/WhatsApp_Chat_with_Michelle_Morris.txt',
    'C:/Users/Dev/Documents/GitHub/Dona_project/Example_chat_data/WhatsApp_Chat_with_Sherry_Flores.txt'
]

# Function to extract names from a single chat file
def extract_names_from_chats(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.readlines()

    names = set()
    for line in data:
        # Extract sender names using regex
        match = re.match(r"\d{2}/\d{2}/\d{4}, \d{2}:\d{2} - (.*?): ", line)
        if match:
            sender = match.group(1)
            names.add(sender)

    return names
    
def identify_donor(file_paths):
    common_names = extract_names_from_chats(file_paths[0])

    for path in file_paths[1:]:
        names = extract_names_from_chats(path)
        common_names = common_names.intersection(names)

    # Return a string instead of a set
    return common_names.pop() if common_names else "No common name found."


# Identify the donor
donor_name = identify_donor(file_paths)
print("Donor's Name:", donor_name)
```

```python
def calculate_word_counts(file_path):
    word_counts = defaultdict(int)
    
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.readlines()

    for line in data:
        match = re.match(r"(\d{2}/\d{2}/\d{4}), (\d{2}:\d{2}) - (.*?): (.*)", line)
        if match:
            sender = match.group(3)
            message = match.group(4)
            word_counts[sender] += len(message.split())

    return word_counts

# Function to calculate donor's percentage contribution in each chat
def calculate_donor_percentage(file_paths, donor_name):
    results = []

    for path in file_paths:
        word_counts = calculate_word_counts(path)
        total_words = sum(word_counts.values())
        donor_words = word_counts.get(donor_name, 0)

        # Calculate percentage
        percentage = (donor_words / total_words) * 100 if total_words > 0 else 0
        results.append({
            "Chat File": os.path.basename(path),
            "Donor Percentage": percentage
        })

    return pd.DataFrame(results)

# Calculating donor's percentage contribution
donor_percentage_df = calculate_donor_percentage(file_paths, donor_name)


plt.figure(figsize=(12, 7))
plt.bar(donor_percentage_df["Chat File"], donor_percentage_df["Donor Percentage"], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title(f"Donor's Word Count Percentage in Each Chat - {donor_name}")
plt.xlabel("Chat File")
plt.ylabel("Percentage (%)")
plt.tight_layout()
plt.show()

print(donor_percentage_df)
```

Burstiness of donor in each chat

```python
from datetime import datetime
from statistics import mean, stdev
import pandas as pd
import matplotlib.pyplot as plt

# Function to extract donor's message timestamps
def extract_donor_timestamps(file_paths, donor_name):
    timestamps = []

    with open(file_paths, 'r', encoding='utf-8') as file:
        for line in file:
            match = re.match(r"(\d{2}/\d{2}/\d{4}), (\d{2}:\d{2}) - (.*?):", line)
            if match:
                date_str, time_str, sender = match.groups()
                if sender == donor_name:
                    timestamp = datetime.strptime(f"{date_str} {time_str}", "%d/%m/%Y %H:%M")
                    timestamps.append(timestamp)

    return timestamps

# Function to calculate burstiness for a list of timestamps
def calculate_burstiness(timestamps):
    if len(timestamps) < 2:
        return None
    gaps = [(timestamps[i+1] - timestamps[i]).total_seconds() / 60.0 for i in range(len(timestamps) - 1)]
    if len(gaps) < 2:
        return None
    mu = mean(gaps)
    sigma = stdev(gaps)
    return (sigma - mu) / (sigma + mu) if (sigma + mu) != 0 else 0

# Function to calculate burstiness per file for the donor
def calculate_donor_burstiness(file_paths, donor_name):
    burstiness_scores = []

    for path in file_paths:
        timestamps = extract_donor_timestamps(path, donor_name)
        burstiness = calculate_burstiness(timestamps)
        burstiness_scores.append({
            "Chat File": os.path.basename(path),
            "Donor Burstiness": burstiness
        })

    return pd.DataFrame(burstiness_scores)

# Calculating and visualizing burstiness
donor_burstiness_df = calculate_donor_burstiness(file_paths, donor_name)

plt.figure(figsize=(12, 7))
plt.bar(donor_burstiness_df["Chat File"], donor_burstiness_df["Donor Burstiness"], color='salmon')
plt.xticks(rotation=45, ha='right')
plt.title(f"Donor's Message Burstiness in Each Chat - {donor_name}")
plt.xlabel("Chat File")
plt.ylabel("Burstiness Score")
plt.tight_layout()
plt.show()

print(donor_burstiness_df)
```

I analyzed the donor's burstiness using time gaps between their own messages. Kyle Adkins consistently exhibits burst-style messaging across all chats, with the highest irregularity observed in his conversation with Dr. Heather Hanson. His least bursty, more steady conversations are with Sherry Flores and Donna Patterson.This suggests varying engagement rhythms depending on the contact, possibly reflecting the nature of those relationships.

Interaction Balance

```python
# Function to count messages for each sender
def count_messages(file_path):
    message_counts = defaultdict(int)

    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            match = re.match(r"\d{2}/\d{2}/\d{4}, \d{2}:\d{2} - (.*?):", line)
            if match:
                sender = match.group(1)
                message_counts[sender] += 1

    return message_counts

# Function to calculate interaction balance for the donor
def calculate_interaction_balance(file_paths, donor_name):
    results = []

    for path in file_paths:
        counts = count_messages(path)
        donor_messages = counts.get(donor_name, 0)

        # only two people in the chat. Identifying the other participant
        other_messages = sum(counts.values()) - donor_messages
        total = donor_messages + other_messages

        balance = (donor_messages - other_messages) / total if total > 0 else 0

        results.append({
            "Chat File": os.path.basename(path),
            "Interaction Balance": balance
        })

    return pd.DataFrame(results)

interaction_balance_df = calculate_interaction_balance(file_paths, donor_name)

plt.figure(figsize=(12, 7))
plt.bar(interaction_balance_df["Chat File"], interaction_balance_df["Interaction Balance"], color='orange')
plt.axhline(0, color='gray', linestyle='--')  # baseline at 0
plt.xticks(rotation=45, ha='right')
plt.title(f"Donor's Interaction Balance in Each Chat - {donor_name}")
plt.xlabel("Chat File")
plt.ylabel("Interaction Balance")
plt.tight_layout()
plt.show()

print(interaction_balance_df)
```

Kyle (the donor) tended to dominate the conversation most with Sherry Flores. Donor was sending a lot more messages. Possibly leading or initiating most of the conversation, followed by Donna Patterson also shows noticeable donor dominance reflects greater effort from donors side. With Michelle Morris, Jeffery Hill, and Dr. Heather Hanson,These are close to 0, indicating a fairly balanced conversation both participants contributed similarly,the interaction was more balanced.

```python
# Function to calculate word counts for each sender
def calculate_word_counts(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.readlines()

    word_counts = defaultdict(int)
    for line in data:
        match = re.match(r"\d{2}/\d{2}/\d{4}, \d{2}:\d{2} - (.*?): (.*)", line)
        if match:
            sender, message = match.groups()
            word_counts[sender] += len(message.split())

    return word_counts

# Function to calculate Gini Index
def calculate_gini_index(word_counts):
    sorted_counts = sorted(word_counts.values())
    n = len(sorted_counts)
    if n == 0:
        return 0
    cumulative_sum = sum((i + 1) * count for i, count in enumerate(sorted_counts))
    gini = (2 * cumulative_sum) / (n * sum(sorted_counts)) - (n + 1) / n
    return gini

```

```python
import os
import re
import pandas as pd
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt

file_paths = [
    'C:/Users/Dev/Documents/GitHub/Dona_project/Example_chat_data/WhatsApp_Chat_with_Donna_Patterson.txt',
    'C:/Users/Dev/Documents/GitHub/Dona_project/Example_chat_data/WhatsApp_Chat_with_Dr._Heather_Hanson.txt',
    'C:/Users/Dev/Documents/GitHub/Dona_project/Example_chat_data/WhatsApp_Chat_with_Jeffery_Hill_.txt',
    'C:/Users/Dev/Documents/GitHub/Dona_project/Example_chat_data/WhatsApp_Chat_with_Michelle_Morris.txt',
    'C:/Users/Dev/Documents/GitHub/Dona_project/Example_chat_data/WhatsApp_Chat_with_Sherry_Flores.txt'
]

# Function to parse WhatsApp chat data and extract potential names
def extract_names_from_chats(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.readlines()

    names = set()
    for line in data:
        # Extracting sender names using regex
        match = re.match(r"\d{2}/\d{2}/\d{4}, \d{2}:\d{2} - (.*?): ", line)
        if match:
            sender = match.group(1)
            names.add(sender)

    return names

# Main function to extract names from all chat files
def identify_person(file_paths):
    potential_names = set()
    for path in file_paths:
        names = extract_names_from_chats(path)
        potential_names.update(names)

    return potential_names

# Extracting and printing the identified names
identified_names = identify_person(file_paths)
print("Identified Names:", identified_names)
```

```python
# Process all chats and calculating Gini Index
results = []
for path in file_paths:
    word_counts = calculate_word_counts(path)
    gini_index = calculate_gini_index(word_counts)
    results.append({
        "Chat File": os.path.basename(path),
        "Gini Index": gini_index
    })

# Creating a DataFrame
gini_df = pd.DataFrame(results)

# Displaying the Gini Index results as a bar chart for better visualization
plt.figure(figsize=(12, 7))
plt.bar(gini_df["Chat File"], gini_df["Gini Index"], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title("Gini Index (Interaction Heterogeneity) for Each Chat")
plt.xlabel("Chat File")
plt.ylabel("Gini Index")
plt.tight_layout()
plt.show()

# Displaying the DataFrame
print(gini_df)
```

What Does the metric tell (Gini Index)?

The Gini Index measures the difference of communication in each chat. A lower value indicates a more balanced interaction where participants contribute similarly, while a higher value signifies that one participant dominates the conversation.

Interesting Observations:

The Gini Index for Dr. Heather Hanson's chat is the lowest (0.009), indicating a balanced interaction pattern.

The highest Gini Index is in the chat with Sherry Flores (0.141), suggesting a more dominant communication style by one participant.

Other participants, such as Jeffery Hill and Michelle Morris, have intermediate Gini values, indicating moderately balanced interactions.
